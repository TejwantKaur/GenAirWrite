```python
import cv2
import time
import os
import HandTrackingModule as htm

wCam, hCam = 640, 480

cap = cv2.VideoCapture(0)
cap.set(3, wCam) # width
cap.set(4, hCam) # height

detector = htm.handDetector(num_hands=1)
```

```python
if not cap.isOpened():
    print("Cannot open camera")
    exit()
    
tipIds = [4,8,12,16,20]
try:
    while True:
        success, img = cap.read()
    
        if not success or img is None:
            print("Camera not working")
            break
        img = detector.findHands(img)
        lmList = detector.findPosition(img, draw=False)
    
        # right hand
        if len(lmList) != 0:
            fingers=[]
            # Thumb
            if lmList[tipIds[0]][1] > lmList[tipIds[0]-1][1]: # thumb kioki thle ni jnda iskrke 
                fingers.append(1)
            else:
                fingers.append(0)
    
            # 4 fingers
            for id in range(1,5):
                # if lmList[8][2] < lmList[6][2]:  #[_,x,y] we are taking y value
                if lmList[tipIds[id]][2] < lmList[tipIds[id]-2][2]:
                    fingers.append(1)
                else:
                    fingers.append(0)
    
            # print(fingers)
            totalFingers = fingers.count(1)
            print(totalFingers)
    
        cv2.imshow("Image", img)
        if cv2.waitKey(1) & 0xFF == 27:  # ESC
            print("Exiting")
            break
finally:
    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```
